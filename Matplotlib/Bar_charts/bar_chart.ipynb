```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]

xpos = np.arange(len(company)) #Company is of string type bar will not take string value it will comnvert it into numbers.
xpos

plt.xticks(xpos, company) #It will replace the xpos value with the company names
plt.ylabel('revenue')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2, profit,width=0.4, label="Profit")
plt.legend(shadow=True)
plt.show()

```

```python
#Horizontal bar Charts
#Xticks used for Vertical bar charts and yticks used for horizontal bar charts
#For Horizontal bar char we need to use plt.barh() function

plt.yticks(xpos, company)#It will replace the xpos value with the company names
plt.xlabel('revenue')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2, revenue,label="Revenue")
plt.barh(xpos+0.2, profit,label="Profit")
plt.legend(shadow=True)
plt.show()
```
