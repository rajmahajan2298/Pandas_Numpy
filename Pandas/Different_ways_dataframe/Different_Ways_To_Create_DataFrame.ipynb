```python
#Different Ways to create DataFrame -

#Using CSV
#Using Excel
#Using Python Dictonary
#Using List of Tuples
#Using List of Dictonary

#There are other many ways to create a DataFrame in Python
```

```python
#Using CSV -

import pandas as pd
df = pd.read_csv('weather_data.csv')
df
```

```python
#Using Excel -

import pandas as pd
df = pd.read_excel('weather_data.xlsx', 'weather_data')
df
```

```python
#Using Python Dictonary - (Every Elements needs to have same size of array)

import pandas as pd

weather_data = {
    'days': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017'],
    'temparture': [32,35,34,36],
    'windspeed': [6,7,2,8],
    'event': ['Rain','Suuny','Snow','storm']
}

df = pd.DataFrame(weather_data)
df
```

```python
#Using List of Tuples - (Needs to assign columns name)

import pandas as pd

weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',34,2,'Snow'),
    ('1/4/2017',36,8,'Strorm')
]

df = pd.DataFrame(weather_data, columns=['days','temperature','windspeed','event'])
df
```

```python
#Using List of Dictonary -

import pandas as pd

weather_data = [
    {'days':'1/1/2017','temperature':32,'Windspeed':6,'event':'Rain'},
    {'days':'1/2/2017','temperature':35,'Windspeed':7,'event':'Sunny'},
    {'days':'1/3/2017','temperature':34,'Windspeed':2,'event':'Snow'},
    {'days':'1/4/2017','temperature':36,'Windspeed':8,'event':'Storm'}
]

df = pd.DataFrame(weather_data)
df
```
